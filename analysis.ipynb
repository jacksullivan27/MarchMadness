```python
season_results = pd.read_csv("march_games/MRegularSeasonCompactResults.csv")
season_wins = season_results.groupby(['Season', 'WTeamID']).size().reset_index(name='Wins')
season_losses = season_results.groupby(['Season', 'LTeamID']).size().reset_index(name='Losses')
season_record = season_wins.merge(season_losses, left_on=['Season', 'WTeamID'], right_on=['Season', 'LTeamID'], how='outer')
season_record['Record'] = season_record['Wins'].fillna(0) / (season_record['Wins'].fillna(0) + season_record['Losses'].fillna(0))
season_record = season_record.merge(tourney_teams[['TeamID', 'TeamName']], left_on='WTeamID', right_on='TeamID')
season_record.drop(columns=['WTeamID', 'LTeamID'], inplace=True)
season_record = season_record[['Season', 'TeamName', 'TeamID', 'Wins', 'Losses', 'Record']]
season_record.head()
```

```python
tourney_games = pd.read_csv("march_games/MNCAATourneyCompactResults.csv")
tourney_games = tourney_games[tourney_games['Season'] >= 2002]
seeds = pd.read_csv("march_games/MNCAATourneySeeds.csv")
tourney_games = tourney_games.merge(tourney_teams[['TeamID', 'TeamName']], left_on='WTeamID', right_on='TeamID').merge(seeds, left_on=['Season', 'WTeamID'], right_on=['Season', 'TeamID'])
tourney_games.drop(columns=['TeamID_x', 'TeamID_y', 'WLoc', 'NumOT'], inplace=True)
tourney_games.rename(columns={'TeamName': 'WTeamName', 'Seed': 'WSeed'}, inplace=True)
tourney_games['WSeed'] = tourney_games['WSeed'].str.extract('(\d+)').astype(int)
tourney_games = tourney_games.merge(season_record, left_on=['Season', 'WTeamID'], right_on=['Season', 'TeamID'], how='left').drop(columns=['Wins', 'Losses', 'TeamID', 'TeamName'])
tourney_games.rename(columns={'Record': 'WRecord'}, inplace=True)
tourney_games = tourney_games.merge(tourney_teams[['TeamID', 'TeamName']], left_on='LTeamID', right_on='TeamID').merge(seeds, left_on=['Season', 'LTeamID'], right_on=['Season', 'TeamID'])
tourney_games.drop(columns=['TeamID_x', 'TeamID_y'], inplace=True)
tourney_games.rename(columns={'TeamName': 'LTeamName', 'Seed': 'LSeed'}, inplace=True)
tourney_games['LSeed'] = tourney_games['LSeed'].str.extract('(\d+)').astype(int)
tourney_games = tourney_games.merge(season_record, left_on=['Season', 'LTeamID'], right_on=['Season', 'TeamID'], how='left').drop(columns=['Wins', 'Losses', 'TeamID', 'TeamName'])
tourney_games.rename(columns={'Record': 'LRecord'}, inplace=True)
tourney_games
```
