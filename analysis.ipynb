```python
tourney_teams = pd.read_csv("march_games/MTeams.csv")
tourney_teams['TeamName'] = tourney_teams['TeamName'].str.replace(' St', ' St.')
tourney_teams['TeamName'] = tourney_teams['TeamName'].map(lambda x: tourney_to_season.get(x, x))
tourney_teams.head()
```

```python
import warnings

season_stat_teams = kenpom[["Season", "TeamName"]]
kenpom['TeamName'] = kenpom['TeamName'].str.replace('Wisconsin Green Bay', 'Green Bay')
kenpom['TeamName'] = kenpom['TeamName'].str.replace('Wisconsin Milwaukee', 'Milwaukee')
kenpom['TeamName'] = kenpom['TeamName'].str.replace('Troy St.', 'Troy')
kenpom['TeamName'] = kenpom['TeamName'].str.replace('Louisiana Lafayette','Louisiana')
kenpom['TeamName'] = kenpom['TeamName'].str.replace('College of Charleston', 'Charleston')
# season_stat_teams['TeamName'] = season_stat_teams['TeamName'].str.replace(" St", " St.")
warnings.filterwarnings('ignore')
season_stat_teams = kenpom[["Season", "TeamName"]]
```

```python
season_results = pd.read_csv("march_games/MRegularSeasonCompactResults.csv")
season_wins = season_results.groupby(['Season', 'WTeamID']).size().reset_index(name='Wins')
season_losses = season_results.groupby(['Season', 'LTeamID']).size().reset_index(name='Losses')
season_record = season_wins.merge(season_losses, left_on=['Season', 'WTeamID'], right_on=['Season', 'LTeamID'], how='outer')
season_record['Record'] = season_record['Wins'].fillna(0) / (season_record['Wins'].fillna(0) + season_record['Losses'].fillna(0))
season_record = season_record.merge(tourney_teams[['TeamID', 'TeamName']], left_on='WTeamID', right_on='TeamID')
season_record.drop(columns=['WTeamID', 'LTeamID'], inplace=True)
season_record = season_record[['Season', 'TeamName', 'TeamID', 'Wins', 'Losses', 'Record']]
season_record.head()
```

```python
tourney_games = pd.read_csv("march_games/MNCAATourneyCompactResults.csv")
tourney_games = tourney_games[tourney_games['Season'] >= 2002]
seeds = pd.read_csv("march_games/MNCAATourneySeeds.csv")
tourney_games = tourney_games.merge(tourney_teams[['TeamID', 'TeamName']], left_on='WTeamID', right_on='TeamID').merge(seeds, left_on=['Season', 'WTeamID'], right_on=['Season', 'TeamID'])
tourney_games.drop(columns=['TeamID_x', 'TeamID_y', 'WLoc', 'NumOT'], inplace=True)
tourney_games.rename(columns={'TeamName': 'WTeamName', 'Seed': 'WSeed'}, inplace=True)
tourney_games['WSeed'] = tourney_games['WSeed'].str.extract('(\d+)').astype(int)
tourney_games = tourney_games.merge(season_record, left_on=['Season', 'WTeamID'], right_on=['Season', 'TeamID'], how='left').drop(columns=['Wins', 'Losses', 'TeamID', 'TeamName'])
tourney_games.rename(columns={'Record': 'WRecord'}, inplace=True)
tourney_games = tourney_games.merge(tourney_teams[['TeamID', 'TeamName']], left_on='LTeamID', right_on='TeamID').merge(seeds, left_on=['Season', 'LTeamID'], right_on=['Season', 'TeamID'])
tourney_games.drop(columns=['TeamID_x', 'TeamID_y'], inplace=True)
tourney_games.rename(columns={'TeamName': 'LTeamName', 'Seed': 'LSeed'}, inplace=True)
tourney_games['LSeed'] = tourney_games['LSeed'].str.extract('(\d+)').astype(int)
tourney_games = tourney_games.merge(season_record, left_on=['Season', 'LTeamID'], right_on=['Season', 'TeamID'], how='left').drop(columns=['Wins', 'Losses', 'TeamID', 'TeamName'])
tourney_games.rename(columns={'Record': 'LRecord'}, inplace=True)
tourney_games
```

```python
def lookup_season_teams(phrase):
    return kenpom[kenpom['TeamName'].str.contains(phrase)][["Season", "TeamName"]].sort_values('Season')

def lookup_tourney_teams(phrase):
    return tourney_games[(tourney_games['WTeamName'].str.contains(phrase)) | (tourney_games['LTeamName'].str.contains(phrase))].sort_values('Season')
```

```python
matchup_stats = (tourney_games.merge(kenpom, left_on=['Season', 'WTeamName'], right_on=['Season', 'TeamName'], how='left')
    .merge(kenpom, left_on=['Season', 'LTeamName'], right_on=['Season', 'TeamName'], suffixes=('_W', '_L'), how='left'))

temp = (matchup_stats.isna().mean() * 100).sort_values(ascending=False)
dropped_columns = temp[temp > 1].index

matchup_stats.drop(columns=dropped_columns, inplace=True)
matchup_stats.drop(columns=['TeamName_W', 'TeamName_L'], inplace=True)
matchup_stats.columns
```

```python
matchup_stats["HigherSeed"] = matchup_stats.apply(lambda row: row['WTeamName'] if (row['WSeed'] < row['LSeed']) or (row['WSeed'] == row['LSeed'] and (row['WRecord'] > row['LRecord'] or (row['WRecord'] == row['LRecord'] and row['WTeamName'] < row['LTeamName']))) else row['LTeamName'], axis=1)
matchup_stats["HigherSeedWin"] = matchup_stats.apply(lambda row: 1 if row['HigherSeed'] == row['WTeamName'] else 0, axis=1)
matchup_stats.head()
```

```python
# Identify winner and loser columns
columns_w = [col for col in matchup_stats.columns if col.endswith('_W')]
columns_l = [col for col in matchup_stats.columns if col.endswith('_L')]

# Convert columns to numeric
matchup_stats[columns_w + columns_l] = matchup_stats[columns_w + columns_l].apply(pd.to_numeric, errors='coerce')

# Create new columns for the differences based on higher and lower seed
for col_w in columns_w:
    col_l = col_w.replace('_W', '_L')
    if col_l in columns_l:
        # Create a boolean mask to check if the winner was the higher seed
        higher_seed_wins = matchup_stats['HigherSeed'] == matchup_stats['WTeamName']
        
        # Compute differences row-wise
        stat_name = col_w.replace('_W', '')  # Remove _W to get the base stat name
        matchup_stats[f'{stat_name}_diff'] = (
            matchup_stats[col_w] - matchup_stats[col_l]
        ).where(higher_seed_wins, matchup_stats[col_l] - matchup_stats[col_w])

# Remove columns that are 100% NaN
remove = matchup_stats.isna().mean() * 100
matchup_stats.drop(columns=remove[remove == 100].index, inplace=True)
columns_w = [col for col in matchup_stats.columns if col.endswith('_W')]
columns_l = [col for col in matchup_stats.columns if col.endswith('_L')]
matchup_stats.drop(columns=columns_w + columns_l, inplace=True)
matchup_stats['higher_seed_num'] = matchup_stats[['WSeed', 'LSeed']].min(axis=1)
matchup_stats['lower_seed_num'] = matchup_stats[['WSeed', 'LSeed']].max(axis=1)
matchup_stats['higher_record'] = matchup_stats.apply(lambda row: row['WRecord'] if row['HigherSeed'] == row['WTeamName'] else row['LRecord'], axis=1)
matchup_stats['lower_record'] = matchup_stats.apply(lambda row: row['WRecord'] if row['HigherSeed'] != row['WTeamName'] else row['LRecord'], axis=1)

display(list(matchup_stats.columns))
matchup_stats
```
